## Phase Space Trajectories for $V(x)= {(x^2-1)}^2 $ : RK-2 Method

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rk2(x0,v0,tau):
    N=10000
    dt= tau/N
    x= np.zeros(N)
    v= np.zeros(N)
    f= np.zeros(N)
    f1= np.zeros(N)
    f2= np.zeros(N)
    x[0]=x0
    v[0]=v0

    for i in range(N-1):
        f1[i]= -4*x[i]*(x[i]**2-1)
        f2[i]= -4*(x[i]+dt)*((x[i]+dt)**2-1)
        f[i]= (f1[i]+f2[i])/2
        
        v[i+1]= v[i]+f[i]*dt
        x[i+1]= x[i]+v[i]*dt
        
    return x,v
     
```

```python
x1,v1=rk2(1,0.1, 100)
x2,v2=rk2 (-1,0.1, 100)
x3,v3=rk2 (1,10, 15)
fig, (ax1, ax2, ax3) = plt.subplots(3,sharex=True)
fig.suptitle('Phase Space Trajectories for $V(x)= {(x^2-1)}^2 $')
ax1.plot(x1,v1)
ax2.plot(x2, v2)
ax3.plot(x3, v3)

plt.setp(ax1, xticks=[0,-1,1,-2,2], xticklabels=['0', '-1','1','-2','2'],
             yticks=[-4,0,4], yticklabels=['-4','0','4'])

plt.setp(ax2, xticks=[0,-1,1,-2,2,], xticklabels=['0', '-1','1','-2','2'],
             yticks=[-4 ,0, 4], yticklabels=['-4','0','4'])

plt.setp(ax3, xticks=[0,-1,1,-2,2,-3,3], xticklabels=['0', '-1','1','-2','2','-3','3'],
             yticks=[-10 ,0, 10], yticklabels=['-10','0','10'])

ax1.set_title("Initial Condition: ($x_0$, $p_0$)= (1, 0.1)")
ax2.set_title("Initial Condition: ($x_0$, $p_0$)=(-1, 0.1)")
ax3.set_title("Initial Condition: ($x_0$, $p_0$)=(1, 10)")

ax1.set_ylabel('Velocity (v)')
ax2.set_ylabel('Velocity (v)')
ax3.set_ylabel('Velocity (v)')
ax3.set_xlabel('Position (v)')
plt.rcParams['figure.figsize']= (15,8)

# plt.savefig(f"Phase-space-Trajectory-{(x0, v0)}.png")


plt.show()
```
