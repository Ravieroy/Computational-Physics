# Hubbard Model(HM)
---
- Formulated by Martin Gutzwiller, Junjiro Kanamori and John Hubbard (1963)


- Extension of Tight Binding Model

- The idea was to tackle the corelated behavior of solids.
    - Itenerant Ferromagntism of transition metals such as Fe and Ni

- It features an additional term U(originally known as "I" until PAM Anderson. This is the energy required, if a pair of electrons occupy the same lattice. (Coulomb Repulsion)


- Some well known features
    - It produces both metallic and insulating state depending on value of U

 - At half-filling, Mott-Hubbard transition is observed(more about this later).This kind of transition could not be understood in terms of conventional band theory.

 - t-J model: High Tc superconductivity 

- Experimental realisations of HM has been realised in cold-atom trapping experiments (Nature 415,39-44(2002)

### Anti-Commutation Relation

$$\{\hat{c_{j\sigma}}, \hat{c^\dagger_{j\sigma^\prime}}\} = \delta_{j,l}\delta_{\sigma,\sigma^\prime}$$

$$\{\hat{c_{j\sigma}}, \hat{c_{j\sigma}}\} = \{\hat{c^\dagger_{j\sigma}}, \hat{c^\dagger_{j\sigma^\prime}}\} =0 $$

### Constructing the model 

- **Kinetics**: We start with a lattice of atoms(sites) on which fermions can move.

- **Interaction**: We add a two particle interaction term modelled by U. The interaction is zero, for zero and one site occupancy but has value U if the site is doubly occupied.

## Hubbard Hamiltonian
    
$$\hat{H} = -t \sum_{\langle j,l \rangle\sigma} (c_{j\sigma}^\dagger c_{l\sigma}) + (c_{l\sigma}^\dagger c_{j\sigma}) + U \sum_{j} n_{j\uparrow}n_{j\downarrow} + \mu \sum_{j}(n_{j\uparrow} + n_{j\downarrow})$$


- The First term is the Kinetic energy. It describes the destruction of an fermion of spin $\sigma$ on site 1 and its creation on site j (or vice-versa). The symbol $\langle j,l \rangle$ emphasizes that hopping is for neaest neighbours. 


- The second term is the interaction energy which adds an energy U, if the lattice is doubly occupied

- The final term is the chemical potential which controls the filling. If there is one fermion per site, then it is called "half-filling"

## Hubbard Hamiltonian (PHS)

$c_{l\sigma}^\dagger \rightarrow (-1)^l c_{l\sigma}$
    
$$\hat{H} = -t \sum_{\langle j,l \rangle\sigma} (c_{j\sigma}^\dagger c_{l\sigma}) + (c_{l\sigma}^\dagger c_{j\sigma}) + U \sum_{j} (n_{j\uparrow}- \frac{1}{2})(n_{j\downarrow}-\frac{1}{2}) + \mu \sum_{j}(n_{j\uparrow} + n_{j\downarrow})$$



### Inorder to understand few things about the Hamiltonian, we will look at the Single-site limit(i.e t=0)

$$\hat{H}_{t=0} = U \sum_{j} (n_{j\uparrow}- \frac{1}{2})(n_{j\downarrow}-\frac{1}{2}) + \mu \sum_{j}(n_{j\uparrow} + n_{j\downarrow})$$

### Possible Configurations and corresponding energy

$$
\begin{align}
|0 \rangle : & \frac{U}{4}&\\
|\uparrow(\downarrow) \rangle : & -\frac{U}{4} - \mu&\\
% |\rangle : & -\frac{U}{4} - \mu&\\
|\uparrow \downarrow\rangle : & \frac{U}{4} - 2\mu
\end{align}
$$


- Using the mentioned eigenvalues , we can construct the partition function $$Z = Tr[e^{-\beta U/4}]$$, and 



The occupation is given, by

$$\rho = \langle n_\uparrow + n_\downarrow \rangle = Z^{-1} Tr[(n_\uparrow + n_\downarrow)e^{-\beta \hat{H}}]$$

# Enough of handwaving, Show me the model !!

### Consider N atoms in the presence of a z-directed magnetic field of strength H. Suppose that all atoms are identical spin- $\frac{1}{2}$ systems. The total energy of the system is writtem as,

$$ E = -J \sum_{\langle i,j\rangle} s_i s_j - \mu H \sum_{i=1,N} s_i$$

- $\langle i,j\rangle$ is sum over nearest neighbours
- J is the exchnage energy 
- $\mu$ is the atomic magnetic moment

## So now what does the model does actually mean?

Since we are only concerned with the nearest neighbour, $-J \sum_{\langle i,j\rangle} s_i s_j$ can be written more explicitly as, 
$$-J \sum_{i =1}^N s_i s_{i+1}$$

- J > 0, Ferromagnetic interaction (i.e parallel neighboring spins is favoured)
- J < 0, Antiferromagnetic interaction (i.e Anti-parallel neighboring spins is favoured)

## Anderson Impurity Model

$$ \hat{H} = \hat{H}_{band} + \hat{H}_{imp} + \hat{H}_{hybd} $$

$$\hat{H}_{band} = \sum_{k,\sigma} \epsilon_k c_{k\sigma}^\dagger c_{k\sigma} $$

$$\hat{H}_{imp} = \sum_{\sigma = \uparrow,\downarrow} \epsilon_0 \hat{n}_{d\sigma} + U \hat{n}_{d\uparrow}\hat{n}_{d\downarrow} $$

$$\hat{H}_{hybd} = \sum_{k,\sigma} t_k c_{d\sigma}^\dagger c_{k\sigma} + h.c $$


where, 

$$\hat{n}_{d\sigma} \equiv c_{d\sigma}^\dagger c_{d\sigma}$$

```python
show_("SIAM_Model.png")
```

## Anderson Impurity States 

$$\hat{H}_{imp} = \sum_{\sigma = \uparrow,\downarrow} \epsilon_0 \hat{n}_{d\sigma} + U \hat{n}_{d\uparrow}\hat{n}_{d\downarrow} $$

$$\hat{n}_{d\uparrow} = \hat{n}_{d\uparrow} = 0 $$

$$E_0 = 0 :  \langle 0 | \sum_{\sigma} \hat{n}_{d\sigma} |0 \rangle = 0$$

$$ \text{Either} \ \hat{n}_{d\uparrow} =0  \ \hat{n}_{d\downarrow} = 1 \ \text{or} \ \hat{n}_{d\uparrow} =1 \ \hat{n}_{d\downarrow} = 0$$ 

$$E_1 = \epsilon_0 :  \langle 1 | \sum_{\sigma} \hat{n}_{d\sigma} |1 \rangle = 1$$



$$ \hat{n}_{d\uparrow} =1  \ \hat{n}_{d\downarrow} = 1 $$

$$E_1 = 2\epsilon_0 + U :  \langle 2 | \sum_{\sigma} \hat{n}_{d\sigma} |2 \rangle = 2$$
