## Anderson Impurity Model

$$ \hat{H} = \hat{H}_{band} + \hat{H}_{imp} + \hat{H}_{hybd} $$

$$\hat{H}_{band} = \sum_{k,\sigma} \epsilon_k c_{k\sigma}^\dagger c_{k\sigma} $$

$$\hat{H}_{imp} = \sum_{\sigma = \uparrow,\downarrow} \epsilon_0 \hat{n}_{d\sigma} + U \hat{n}_{d\uparrow}\hat{n}_{d\downarrow} $$

$$\hat{H}_{hybd} = \sum_{k,\sigma} t_k c_{d\sigma}^\dagger c_{k\sigma} + h.c $$

where, 

$$\hat{n}_{d\sigma} \equiv c_{d\sigma}^\dagger c_{d\sigma}$$

We have, 

$$\langle \hat{n}_\uparrow(\downarrow) \rangle = \frac{1}{2} - \frac{1}{\pi} tan^{-1} \left(\frac{\bar{\epsilon_0 + U\langle \hat{n}_\uparrow(\downarrow)\rangle}}{\Delta}\right)$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calcuate_n_up(n_dn_old,e0,umax=5,delta = 0.1):
    """
    Calculates the average of n_up spin
    """
    return 0.5 - (1/np.pi) * np.arctan((e0 + (umax*n_dn_old))/delta)

def calcuate_n_dn(n_up_old,e0,umax=5,delta = 0.1):
    """
    Calculates the average of n_up spin
    """
    return 0.5 - (1/np.pi) * np.arctan((e0 + (umax*n_up_old))/delta)
```

```python
def create_spins_list(nup_i = 0.5, ndn_i = 0.5, delta = 0.1, umax = 5):
    """
    Creates a list of n_up spins, n_dn spins and n_tot for the energy range 
    
    Arguments:
    nup_i = Initial value for n_up
    ndn_i = Initial value for n_dn
    delta = broadening
    umax = maximum U 
    """
    e0array = np.linspace(-1.5*umax, umax, 51)
    n_up_list = []
    n_dn_list = []
    for e0 in e0array:
        n_up = nup_i
        n_dn = ndn_i
        n_up_old = n_up
        n_dn_old = n_dn
        num1 = n_up
        ndm1 = n_dn
        for i in range(101):
            n_up = calcuate_n_up(n_dn_old,e0)
            n_dn = calcuate_n_dn(n_up_old,e0)
            num1 = n_up_old
            ndm1 = n_dn_old
            n_up_old = n_up
            n_dn_old = n_dn

        n_up_list.append(0.5*(n_up + num1))
        n_dn_list.append(0.5*(n_dn + ndm1))
        
    n_tot = [sum(i) for i in zip(n_up_list, n_dn_list)]
    
    return n_up_list, n_dn_list, n_tot, e0array
    
```

```python
def show_plot():
    """
    Shows the plot of occupation of spins in a range
    """
    plt.figure(figsize=(8,4), tight_layout=True)
    plt.plot(e0array,n_up_list,'^-',label=r"$\langle n_\uparrow \rangle$")
    plt.plot(e0array,n_dn_list,'v-',label=r"$\langle n_\downarrow \rangle$")
    plt.plot(e0array,n_tot,label=r"$\langle n_{tot} \rangle$")
    plt.xlabel("$\epsilon_0$")
    plt.ylabel(r"$\langle n \rangle$")
    plt.grid()
    plt.legend()
    plt.show()
```

```python
def plot_spin_diff():
    
    """
    Shows the plot of spin polarization <n_up> - <n_dn>
    """
    n_diff = [abs(a_i - b_i) for a_i, b_i in zip(n_up_list, n_dn_list)]
    plt.figure(figsize=(8,4), tight_layout=True)
    plt.plot(e0array,n_diff,'r-o')
    plt.xlabel("$\epsilon_0$")
    plt.ylabel(r"$\langle n\uparrow \rangle$ - $\langle n\downarrow \rangle$")
    plt.xlim([-5,0])
    plt.title("Spontaneous Symmetry Breaking")
    plt.grid()
    plt.show()
```

```python
n_up_list, n_dn_list, n_tot,e0array = create_spins_list(nup_i=0.5, ndn_i=0.5) # create a list for up/down/total spins
#use nup_i=0.6, ndn_i=0.4
show_plot()
plot_spin_diff()
```
