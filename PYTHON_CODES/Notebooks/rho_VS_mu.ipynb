```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate_density(mu,T, U=4):
    beta = 1/T
    E0 = (-beta * (U/4))
    E1 = (-beta * (-U/4 - mu))
    E2 = (-beta * (U/4 - 2*mu))
    Z = np.exp(E0) + 2 * np.exp(E1) + np.exp(E2)
    rho = (1/Z) * (2 * np.exp(E1) + 2 * np.exp(E2))
    return rho
```

```python
T = [2,1.25,1,0.5,0.25]
list_ = []
for val in T :
    mu = np.linspace(-6,6,100)
    rho_list = []
    for vals in mu:
        rho = calculate_density(vals,val)
        rho_list.append(rho)
    list_.append(rho_list)
```

```python
def show():
    for i in range(len(list_)):
        plt.plot(mu,list_[i],label=f"T : {T[i]}K")
        plt.grid()
        plt.xlabel("$\mu$")
        plt.ylabel(r"$\rho$")
        plt.xlim([-6,6])
        plt.ylim([-0.05,2.05])
        plt.legend()
        plt.title(r"$\rho$ VS $\mu$ | t : 0 |  U : 4")
        plt.axhline(1, linestyle='--')
        plt.axvline(0, linestyle='--')
        plt.axvline(-2, linestyle=':',color='r')
        plt.axvline(2, linestyle=':',color='r')
        plt.savefig('rho_VS_mu.png', dpi = 100)
    return plt.show()
```

```python
show()
```
